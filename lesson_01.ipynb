## 01.02 Exercise / Exercício:

**EN:** Define a function factorial that takes an integer x as input. Calculate and return the factorial of that number.

**PT:** Defina uma função factorial que toma um número inteiro x como entrada. Calcule e retorne o fatorial daquele número.

```python
def factorial(x):
    if x == 0:
        print("0! = 1")
        return 1

    resultado = 1
    passos = []  # lista para guardar os números que vamos multiplicar

    for i in range(x, 0, -1):  # começa de x e vai até 1
        resultado *= i
        passos.append(str(i))  # transformar número em string para print

    # criar a string tipo 4*3*2*1
    calculo = "*".join(passos)

    print(f"{x}! = {calculo} = {resultado}")
    return resultado

numero = int(input("Digite um número inteiro: "))
factorial(numero)
```

## 01.03 Exercise / Exercício:

**EN:** Define a function called isnot_prime that takes a number x as input. If it is prime, return False. If it is not prime, return True.

**PT:** Defina uma função chamada isnot_prime que toma um número x como entrada. Se for primo, retorne False. Se não for primo, retorne True.

```python
def isnot_prime(x):
    if x < 2:  # números menores que 2 não são primos
        return True

    for i in range(2, x):
        if x % i == 0:
            return True  # não é primo
    return False  # é primo

numero = int(input("Digite um número inteiro: "))
print("Não é primo?", isnot_prime(numero))
```

**EN:** Write a function called censor that takes two strings, text and word, as input. It should return the text with the chosen word replaced by asterisks. Assume input strings contain no punctuation or uppercase letters. The number of asterisks must match the number of letters in the censored word.

Suponha que as strings de entrada não contêm pontuação ou letras maiúsculas. O número de asteriscos deve corresponder ao número de letras na palavra censurada.

```python
def censor(text, word):
    # cria a versão censurada da palavra usando asteriscos
    censurada = "*" * len(word)

    # substitui todas as ocorrências da palavra no texto
    texto_final = text.replace(word, censurada)

    return texto_final

texto = input("Digite um texto: ")
palavra = input("Digite a palavra a ser censurada: ")

print(censor(texto, palavra))
```

```python
def remove_duplicates(lista):
    # nova lista para armazenar elementos únicos
    nova_lista = []

    # percorre cada elemento da lista original
    for elemento in lista:
        if elemento not in nova_lista:
            nova_lista.append(elemento)  # adiciona apenas se ainda não estiver na nova lista

    return nova_lista

entrada = [1, 2, 2, 3, 4, 4, 5]
print(remove_duplicates(entrada))
```

## 02.03 Exercise / Exercício:

**EN:** Write a function called median that takes a list as input and returns the median value. If the list has an even number of elements, return the average of the two middle elements.

**PT:** Escreva uma função chamada median que tome uma lista como entrada e retorne o valor da mediana. Se a lista contiver um número par de elementos, sua função deve retornar a média dos dois elementos do meio.

```python
def median(lista):
    # cria uma cópia da lista e ordena
    ordenada = sorted(lista)
    n = len(ordenada)
    meio = n // 2

    # se a quantidade de elementos for ímpar, retorna o do meio
    if n % 2 == 1:
        return ordenada[meio]
    else:
        # se for par, retorna a média dos dois do meio
        return (ordenada[meio - 1] + ordenada[meio]) / 2

numeros = [5, 2, 9, 1, 7]
print(median(numeros))
```

## 03.04 Exercise / Exercício:

**EN:** Given a 3x3 square matrix A: 1) Display the matrix line by line. 2) Create a function to calculate the transpose of A. 3) Create a function to verify if A is an Identity matrix. 4) Given another matrix B, calculate A + B.

**PT:** Dada uma matriz quadrada A (3x3): 1) Mostrar no console a matriz, linha a linha. 2) Criar uma função que calcule a transposta de A. 3) Criar uma função que verifique se A é igual à identidade. 4) Dada outra matriz B, calcular A + B.

```python
# Matriz A (3x3)
A = [
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
]

# Mostrar matriz A
print("Matriz A:")
for linha in A:
    print(linha)

# Função para transposta
def transposta(matriz):
    trans = []
    for i in range(len(matriz)):
        nova_linha = []
        for j in range(len(matriz)):
            nova_linha.append(matriz[j][i])
        trans.append(nova_linha)
    return trans

# Função para verificar identidade
def is_identidade(matriz):
    n = len(matriz)
    for i in range(n):
        for j in range(n):
            if i == j and matriz[i][j] != 1:
                return False
            elif i != j and matriz[i][j] != 0:
                return False
    return True

# Função para somar matrizes
def soma_matrizes(M1, M2):
    n = len(M1)
    resultado = []
    for i in range(n):
        linha = []
        for j in range(n):
            linha.append(M1[i][j] + M2[i][j])
        resultado.append(linha)
    return resultado

# Mostrar transposta de A
print("\nTransposta de A:")
for linha in transposta(A):
    print(linha)

# Verificar se A é identidade
print("\nA é identidade?", is_identidade(A))

# Matriz B
B = [
    [2, 1, 0],
    [0, 3, 1],
    [1, 0, 2]
]

# Mostrar A + B
print("\nA + B:")
for linha in soma_matrizes(A, B):
    print(linha)
```

**PT:**
**Crie uma classe Conta:**
* Ao criar uma conta, não há nenhuma transação e o saldo é zero.
* Para criar uma conta, é preciso informar o nome do proprietário e a data de abertura
* Crie as funções sacar e depositar. A função sacar não pode permitir saldo negativo.
* Crie uma função que retorne o número de transações efetuadas e o saldo.
* Crie uma subclasse ContaEspecial que herde de Conta, mas tenha um limite. Permita saques dentro desse limite.
* Crie contas e transações para testar a classe.

```python
# Classe Conta
class Conta:
    def __init__(self, nome, data_abertura):
        self.nome = nome
        self.data_abertura = data_abertura
        self.saldo = 0
        self.transacoes = 0

    # depositar valor
    def depositar(self, valor):
        if valor > 0:
            self.saldo += valor
            self.transacoes += 1

    # sacar valor, sem permitir saldo negativo
    def sacar(self, valor):
        if 0 < valor <= self.saldo:
            self.saldo -= valor
            self.transacoes += 1
        else:
            print("Saldo insuficiente!")

    # retornar saldo e número de transações
    def info(self):
        return f"Saldo: {self.saldo}, Transações: {self.transacoes}"

# Subclasse ContaEspecial com limite de saque
class ContaEspecial(Conta):
    def __init__(self, nome, data_abertura, limite):
        super().__init__(nome, data_abertura)
        self.limite = limite

    # sacar com limite permitido
    def sacar(self, valor):
        if 0 < valor <= self.saldo + self.limite:
            self.saldo -= valor
            self.transacoes += 1
        else:
            print("Valor excede saldo + limite!")

# Testando as classes
# Conta normal
conta1 = Conta("Alice", "2026-03-09")
conta1.depositar(500)
conta1.sacar(200)
conta1.sacar(400)  # não permite
print(conta1.info())

# Conta especial com limite 300
conta2 = ContaEspecial("Bruno", "2026-03-09", 300)
conta2.depositar(500)
conta2.sacar(700)  # permitido pelo limite
conta2.sacar(200)  # excede saldo+limite
print(conta2.info())
```
