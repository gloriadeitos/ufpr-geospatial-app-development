# Exemplo 1 - Criando um novo mapa

Agora vamos começar a trabalhar com a biblioteca Folium. É uma distribuição do Leaflet para criação de mapas web via Python.

Mais informações:
[link text]https://python-visualization.github.io/folium/


```
#importando a biblioteca
import folium

#Para criar um mapa é bem simples, basta definir as coordenadas do centro, as tiles (mapa raster) de fundo e o zoom inicial
m = folium.Map (location = [-25.5,-49.3],
               tiles = 'OpenStreetMap',
               zoom_start =  12,
               attr = 'Colaboradores do OpenStreetMap e Alunos de CartoUFPR/ Dados IPPUC'
               )

# e chamar o mapa aqui
m

```

Outras opções de tiles: https://python-graph-gallery.com/288-map-background-with-folium/

## Vamos agora adicionar um ponto ao mapa:

```
import folium
# esse texto é mostrado num balão ao passar o mouse sobre o ponto
tooltip = 'Clique aqui!'

folium.Marker([-25.5,-49.3], popup='<b>Curitba</b><br><img src="http://multimidia.curitiba.pr.gov.br/2018/capa/00239696.jpg" style="width:128px;height:128px;"><br><a href="http://www.curitiba.pr.gov.br">Site da Prefetura</a> ',
              tooltip=tooltip,
             icon = folium.Icon(icon='star', color = 'green'),
             ).add_to(m)
m.save('/content/drive/My Drive/AAAARafael/meumapa.html ') # abra e observe este arquivo
m
```

Outras opções de ícones: https://getbootstrap.com/docs/3.3/components/

## Agora, um ponto com ícone customizado

```
#chamando o ícone
logo_url = 'http://vignette3.wikia.nocookie.net/the-adventures-of-the-gladiators-of-cybertron/images/6/67/Iron_Man.png/revision/latest?cb=20140720074813'

#coordenadas do poli
UFPR_lat = -25.45
UFPR_lon = -49.23

#ícone customizado
icon = folium.features.CustomIcon(logo_url,\
                                  icon_size=(100, 100))
#adiciona ao mapa
folium.Marker([UFPR_lat, UFPR_lon],
          popup='Homem de Ferro no poli',
          icon=icon).add_to(m)
m
```
